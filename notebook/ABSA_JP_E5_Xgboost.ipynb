```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings, re, optuna, os, requests
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, precision_recall_fscore_support, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import xgboost as xgb
import shap

warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import matplotlib
import os

# Download Noto Sans JP font if not exists (Colab/Kaggle)
font_url = "https://github.com/googlefonts/noto-cjk/raw/main/Sans/OTF/Japanese/NotoSansCJKjp-Regular.otf"
font_path = "/usr/share/fonts/truetype/noto/NotoSansCJKjp-Regular.otf"

if not os.path.exists(font_path):
    os.makedirs(os.path.dirname(font_path), exist_ok=True)
    import requests
    r = requests.get(font_url)
    with open(font_path, "wb") as f:
        f.write(r.content)

# Register and set as default
matplotlib.font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Noto Sans CJK JP'
plt.rcParams['axes.unicode_minus'] = False  # For minus sign

# Check (optional)
# print([f.name for f in matplotlib.font_manager.fontManager.ttflist if 'Noto' in f.name])


class JapaneseBusinessInsightExtractor:
    def __init__(self):
        self.aspects = {
            'fit': ['フィット', 'サイズ', 'きつい', 'ゆるい', '小さい', '大きい', 'ちっちゃい', 'でかい', '合う', '合わない', 'ぴったり', 'ぶかぶか'],
            'quality': ['品質', '質', '素材', 'クオリティ', '安っぽい', '高級', '作り', '丈夫', 'しっかり', 'ちゃちい', 'ペラペラ', '生地'],
            'style': ['スタイル', 'デザイン', '色', 'おしゃれ', 'かわいい', 'かっこいい', 'ダサい', 'イケてる', 'オシャレ', '可愛い'],
            'comfort': ['快適', '心地', '着心地', '楽', 'らく', 'きもちいい', '気持ちいい', 'ふわふわ', 'さらさら', 'ごわごわ'],
            'price': ['価格', '値段', '金額', '高い', '安い', 'コスパ', 'お得', 'もったいない', 'ぼったくり', '激安', 'プチプラ'],
            'shipping': ['配送', '発送', '届く', '配達', '到着', '早い', '遅い', '梱包', 'はやい', 'おそい', 'すぐ']
        }
        self.positive_words = ['最高', '素晴らしい', '良い', 'いい', '完璧', '最強', 'やばい', 'すごい', 'めっちゃいい', 'マジいい', 'ヤバい', 'エモい', 'しあわせ', '幸せ']
        self.negative_words = ['最悪', 'ひどい', '悪い', 'だめ', 'クソ', 'うざい', 'ムカつく', 'イライラ', 'がっかり', 'ショック', 'つまらない', 'びみょう', '微妙']
        self.neutral_words = ['普通', 'まあまあ', '平均', 'ふつう', 'そこそこ', 'まずまず', '標準', 'ありきたり', 'こんなもん']

    def generate_business_recommendations(self, df):
        print("\n=== ビジネス分析と推奨事項 ===")
        for aspect in self.aspects.keys():
            aspect_col = f'aspect_{aspect}'
            if aspect_col in df.columns:
                aspect_mentions = df[aspect_col].sum()
                negative_mentions = df[(df[aspect_col]) & (df['sentiment'] == 0)].shape[0]
                neutral_mentions = df[(df[aspect_col]) & (df['sentiment'] == 1)].shape[0]
                if aspect_mentions > 0:
                    negative_rate = negative_mentions / aspect_mentions * 100
                    neutral_rate = neutral_mentions / aspect_mentions * 100
                    print(f"• {aspect.upper()}: {aspect_mentions} 件の言及, {negative_rate:.1f}% ネガティブ, {neutral_rate:.1f}% 中立")
        return {"status": "completed"}

def executive_summary(df):
    print("\n=== エグゼクティブサマリー ===")
    total_reviews = len(df)
    sentiment_dist = df['sentiment'].value_counts(normalize=True) * 100
    sentiment_labels = {0: 'ネガティブ', 1: '中立', 2: 'ポジティブ'}
    print(f"分析したレビュー総数: {total_reviews:,}")
    print("感情分布:")
    for sentiment, pct in sentiment_dist.items():
        print(f"  • {sentiment_labels.get(sentiment, sentiment)}: {pct:.1f}%")

def plot_overall_sentiment(df):
    plt.figure(figsize=(15, 5))
    plt.rcParams['font.family'] = 'DejaVu Sans'  # For Japanese text display
    plt.subplot(1, 3, 1)
    sentiment_counts = df['sentiment'].value_counts()
    sentiment_labels = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    labels = [sentiment_labels.get(k, str(k)) for k in sentiment_counts.index]
    colors = ['#ff6b6b', '#ffd93d', '#6bcf7f']
    plt.pie(sentiment_counts.values, labels=labels, autopct='%1.1f%%',
            colors=colors, startangle=90)
    plt.title('Sentiment Distribution')
    plt.subplot(1, 3, 2)
    plt.hist(df['sentiment'], bins=3, color='skyblue', alpha=0.7, edgecolor='black')
    plt.xlabel('Sentiment')
    plt.ylabel('Frequency')
    plt.title('Sentiment Distribution')
    plt.xticks([0, 1, 2], ['Negative', 'Neutral', 'Positive'])
    plt.subplot(1, 3, 3)
    sentiment_counts_df = df['sentiment'].value_counts().reindex([0, 1, 2], fill_value=0)
    plt.bar(range(3), sentiment_counts_df.values, color=colors, alpha=0.7)
    plt.title('Sentiment Counts')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.xticks(range(3), ['Negative', 'Neutral', 'Positive'])
    plt.tight_layout()
    plt.show()

def plot_negative_aspect_breakdown(df):
    aspects = ['fit', 'quality', 'style', 'comfort', 'price', 'shipping']
    negative_counts, neutral_counts = [], []
    for aspect in aspects:
        aspect_col = f'aspect_{aspect}'
        if aspect_col in df.columns:
            neg_count = df[(df[aspect_col]) & (df['sentiment'] == 0)].shape[0]
            neu_count = df[(df[aspect_col]) & (df['sentiment'] == 1)].shape[0]
            negative_counts.append(neg_count)
            neutral_counts.append(neu_count)
        else:
            negative_counts.append(0)
            neutral_counts.append(0)
    plt.figure(figsize=(12, 6))
    x = np.arange(len(aspects))
    width = 0.35
    plt.bar(x - width/2, negative_counts, width, label='Negative', color='#ff6b6b', alpha=0.7)
    plt.bar(x + width/2, neutral_counts, width, label='Neutral', color='#ffd93d', alpha=0.7)
    plt.title('Negative and Neutral Sentiment by Aspect')
    plt.xlabel('Aspects')
    plt.ylabel('Number of Reviews')
    plt.xticks(x, aspects, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_aspect_sentiment_heatmap(df):
    aspects = ['fit', 'quality', 'style', 'comfort', 'price', 'shipping']
    sentiments = [0, 1, 2]
    sentiment_labels = ['Negative', 'Neutral', 'Positive']
    heatmap_data = []
    for aspect in aspects:
        aspect_col = f'aspect_{aspect}'
        row = []
        for sentiment in sentiments:
            if aspect_col in df.columns:
                count = df[(df[aspect_col]) & (df['sentiment'] == sentiment)].shape[0]
                row.append(count)
            else:
                row.append(0)
        heatmap_data.append(row)
    plt.figure(figsize=(10, 6))
    sns.heatmap(heatmap_data,
                xticklabels=sentiment_labels,
                yticklabels=aspects,
                annot=True,
                fmt='d',
                cmap='RdYlBu_r',
                cbar_kws={'label': 'Number of Reviews'})
    plt.title('Aspect-Sentiment Heatmap')
    plt.xlabel('Sentiment')
    plt.ylabel('Aspects')
    plt.tight_layout()
    plt.show()

class JapaneseABSAXGBoostTabularPipeline:
    def __init__(self, max_train_samples=8000, max_test_samples=2000, n_splits=5, tfidf_max_features=200):
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.insight_extractor = JapaneseBusinessInsightExtractor()
        self.max_train_samples = max_train_samples
        self.max_test_samples = max_test_samples
        self.n_splits = n_splits
        self.feature_cols = []
        self.tfidf_max_features = tfidf_max_features
        self.tfidf_vec = TfidfVectorizer(max_features=tfidf_max_features, token_pattern=r'\b\w+\b')
        
    def tokenize_japanese(self, text):
        # Use MeCab if available, else char-level tokenization
        return ' '.join(list(text))

    def load_and_preprocess_data(self):
        print("WRIME データセットを読み込み中...")
        wrime_url = "https://raw.githubusercontent.com/ids-cv/wrime/refs/heads/master/wrime-ver1.tsv"
        wrime_path = "/kaggle/working/wrime-ver1.tsv"
        try:
            if not os.path.exists(wrime_path):
                print("WRIME データセットをダウンロード中...")
                r = requests.get(wrime_url)
                open(wrime_path, "wb").write(r.content)
            df_wrime = pd.read_csv(wrime_path, sep="\t")
            df_wrime = df_wrime.dropna(subset=["Sentence"])
            def get_sentiment(row):
                if row['Avg. Readers_Joy'] > max(row['Avg. Readers_Sadness'], row['Avg. Readers_Anger']):
                    return 2
                elif (row['Avg. Readers_Sadness'] > row['Avg. Readers_Joy']) or (row['Avg. Readers_Anger'] > row['Avg. Readers_Joy']):
                    return 0
                else:
                    return 1
            df_wrime['sentiment'] = df_wrime.apply(get_sentiment, axis=1)
            df_wrime = df_wrime.rename(columns={'Sentence': 'review_text'})
            df_wrime['rating'] = df_wrime['sentiment'] + 1
            print(f"WRIME データセットから {len(df_wrime)} 件のレビューを読み込みました")
        except Exception as e:
            print(f"データセット読み込みエラー: {e}")
            print("合成日本語データを使用します...")
            return self._create_synthetic_japanese_data()
        train_df, test_df = train_test_split(df_wrime, test_size=0.2, random_state=42, stratify=df_wrime['sentiment'])
        train_df = train_df.head(self.max_train_samples)
        test_df = test_df.head(self.max_test_samples)
        return train_df, test_df

    def _create_synthetic_japanese_data(self):
        print("合成日本語レビューデータを作成中...")
        np.random.seed(42)
        pos_templates = [
            "この{item}は{aspect}が{positive_word}です！",
            "{item}の{aspect}がマジで{positive_word}！",
            "めっちゃ{positive_word}な{item}、{aspect}も最高",
            "{positive_word}すぎる{item}！{aspect}がヤバい",
            "この{item}、{aspect}が{positive_word}でおすすめ"
        ]
        neg_templates = [
            "この{item}は{aspect}が{negative_word}です",
            "{item}の{aspect}がクソ{negative_word}",
            "マジで{negative_word}な{item}、{aspect}も最悪",
            "{negative_word}すぎる{item}…{aspect}がダメ",
            "この{item}、{aspect}が{negative_word}でがっかり"
        ]
        neu_templates = [
            "この{item}は{aspect}が普通です",
            "{item}の{aspect}はまあまあかな",
            "普通の{item}、{aspect}は可もなく不可もなく",
            "この{item}、{aspect}はそこそこ",
            "{item}の{aspect}は標準的"
        ]
        items = ["服", "トップス", "ブラウス", "セーター", "ジャケット", "パンツ", "スカート", "ワンピース"]
        aspects = ["フィット感", "品質", "デザイン", "着心地", "素材", "色"]
        pos_words = ["良い", "素晴らしい", "最高", "完璧", "いい感じ", "エモい"]
        neg_words = ["悪い", "ひどい", "最悪", "ダメ", "微妙", "イマイチ"]
        n_samples = self.max_train_samples + self.max_test_samples
        reviews, sentiments = [], []
        for i in range(n_samples):
            rand = np.random.random()
            if rand < 0.6:  # positive
                review = np.random.choice(pos_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects),
                    positive_word=np.random.choice(pos_words)
                )
                sentiment = 2
            elif rand < 0.8:  # neutral
                review = np.random.choice(neu_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects)
                )
                sentiment = 1
            else:  # negative
                review = np.random.choice(neg_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects),
                    negative_word=np.random.choice(neg_words)
                )
                sentiment = 0
            reviews.append(review)
            sentiments.append(sentiment)
        df = pd.DataFrame({'review_text': reviews, 'sentiment': sentiments})
        df['rating'] = df['sentiment'] + 1
        train_df = df.iloc[:self.max_train_samples].copy()
        test_df = df.iloc[self.max_train_samples:self.max_train_samples + self.max_test_samples].copy()
        return train_df, test_df

    def extract_comprehensive_features(self, df, fit_tfidf=False):
        print("包括的特徴量とTF-IDFを抽出中...")
        aspects = self.insight_extractor.aspects
        for aspect, keywords in aspects.items():
            pattern = '|'.join([re.escape(kw) for kw in keywords])
            df[f'aspect_{aspect}'] = df['review_text'].str.contains(pattern, na=False, regex=True).astype(int)
        df['text_length'] = df['review_text'].str.len()
        df['char_count'] = df['review_text'].str.len()
        df['hiragana_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[あ-ん]', x)))
        df['katakana_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[ア-ン]', x)))
        df['kanji_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[一-龯]', x)))
        df['sentence_count'] = df['review_text'].str.count(r'[。！？]') + 1
        df['exclamation_count'] = df['review_text'].str.count('！') + df['review_text'].str.count(r'\!')
        df['question_count'] = df['review_text'].str.count('？') + df['review_text'].str.count(r'\?')
        df['small_tsu_count'] = df['review_text'].str.count('っ')
        df['elongated_vowel_count'] = df['review_text'].str.count('ー')
        positive_words = self.insight_extractor.positive_words
        negative_words = self.insight_extractor.negative_words
        neutral_words = self.insight_extractor.neutral_words
        def count_sentiment_words(text, word_list):
            return sum(1 for word in word_list if word in text)
        df['positive_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, positive_words))
        df['negative_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, negative_words))
        df['neutral_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, neutral_words))
        for aspect in aspects.keys():
            aspect_col = f'aspect_{aspect}'
            if aspect_col in df.columns:
                df[f'{aspect}_positive_interaction'] = df[aspect_col] * df['positive_word_count']
                df[f'{aspect}_negative_interaction'] = df[aspect_col] * df['negative_word_count']
        df['balanced_sentiment'] = ((df['positive_word_count'] > 0) & (df['negative_word_count'] > 0)).astype(int)
        df['short_review'] = (df['text_length'] <= 10).astype(int)
        df['neutral_indicators'] = df['neutral_word_count']
        if fit_tfidf:
            tokenized_texts = df['review_text'].apply(self.tokenize_japanese)
            tfidf_matrix = self.tfidf_vec.fit_transform(tokenized_texts)
        else:
            tokenized_texts = df['review_text'].apply(self.tokenize_japanese)
            tfidf_matrix = self.tfidf_vec.transform(tokenized_texts)
        tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), 
                               columns=[f"tfidf_{i}" for i in range(self.tfidf_max_features)], 
                               index=df.index)
        df = pd.concat([df, tfidf_df], axis=1)
        exclude_cols = ['review_text', 'sentiment', 'rating']
        potential_features = [
            col for col in df.columns
            if col not in exclude_cols and
               not col.startswith("tfidf_") and
               not col.startswith("emb_") and
               not col.startswith("Avg.") and
               not col.startswith("high_")
        ]
        numeric_features = [col for col in potential_features if pd.api.types.is_numeric_dtype(df[col])]
        self.feature_cols = numeric_features
        return df

    def compute_sample_weights(self, y):
        from sklearn.utils.class_weight import compute_class_weight
        classes = np.unique(y)
        class_weights = compute_class_weight('balanced', classes=classes, y=y)
        neutral_class_idx = np.where(classes == 1)[0]
        if len(neutral_class_idx) > 0:
            class_weights[neutral_class_idx[0]] *= 1.5
        weights = np.zeros_like(y, dtype=float)
        for idx, cls in enumerate(classes):
            weights[y == cls] = class_weights[idx]
        return weights

    def fit(self):
        print("日本語ABSA XGBoost (表形式特徴量のみ) パイプラインを開始...")
        train_df, test_df = self.load_and_preprocess_data()
        train_df = self.extract_comprehensive_features(train_df, fit_tfidf=True)
        test_df = self.extract_comprehensive_features(test_df, fit_tfidf=False)
        executive_summary(train_df)
        plot_overall_sentiment(train_df)
        plot_negative_aspect_breakdown(train_df)
        plot_aspect_sentiment_heatmap(train_df)
        self.insight_extractor.generate_business_recommendations(train_df)
        y = train_df['sentiment'].values
        y_test = test_df['sentiment'].values
        X = train_df[self.feature_cols].values
        X_test = test_df[self.feature_cols].values
        X = self.scaler.fit_transform(X)
        X_test = self.scaler.transform(X_test)
        feature_names = self.feature_cols
        skf = StratifiedKFold(n_splits=self.n_splits, shuffle=True, random_state=42)
        train_idx, val_idx = next(skf.split(X, y))
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        weights_train = self.compute_sample_weights(y_train)
        def objective(trial):
            param = {
                "objective": "multi:softprob",
                "num_class": 3,
                "eval_metric": "mlogloss",
                "max_depth": trial.suggest_int("max_depth", 3, 8),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
                "n_estimators": trial.suggest_int("n_estimators", 100, 500),
                "subsample": trial.suggest_float("subsample", 0.7, 1.0),
                "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0),
                "gamma": trial.suggest_float("gamma", 0, 5),
                "lambda": trial.suggest_float("lambda", 1e-3, 10.0, log=True),
                "alpha": trial.suggest_float("alpha", 1e-3, 10.0, log=True),
                "min_child_weight": trial.suggest_int("min_child_weight", 1, 7),
                "tree_method": "auto",
                "use_label_encoder": False,
                "verbosity": 0,
                "random_state": 42
            }
            model = xgb.XGBClassifier(**param)
            model.fit(X_train, y_train, sample_weight=weights_train)
            preds = np.argmax(model.predict_proba(X_val), axis=1)
            _, _, f1, _ = precision_recall_fscore_support(y_val, preds, average='weighted', zero_division=0)
            return 1.0 - f1
        print("Optuna チューニング (重み付きF1スコアを最適化)...")
        study = optuna.create_study(direction="minimize")
        study.optimize(objective, n_trials=20, show_progress_bar=False)
        best_params = study.best_trial.params
        best_params.update({
            "objective": "multi:softprob",
            "num_class": 3,
            "use_label_encoder": False,
            "tree_method": "auto",
            "random_state": 42
        })
        print("最適パラメータ:", best_params)
        oof_preds = np.zeros((len(y), 3))
        cv_acc, cv_prec, cv_rec, cv_f1 = [], [], [], []
        cv_neutral_f1 = []
        for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
            print(f"Fold {fold+1}/{self.n_splits}...")
            X_tr, X_val = X[train_idx], X[val_idx]
            y_tr, y_val = y[train_idx], y[val_idx]
            weights = self.compute_sample_weights(y_tr)
            model = xgb.XGBClassifier(**best_params)
            model.fit(X_tr, y_tr, sample_weight=weights)
            val_pred = np.argmax(model.predict_proba(X_val), axis=1)
            oof_preds[val_idx] = model.predict_proba(X_val)
            acc = accuracy_score(y_val, val_pred)
            prec, rec, f1, _ = precision_recall_fscore_support(y_val, val_pred, average='weighted', zero_division=0)
            prec_per_class, rec_per_class, f1_per_class, _ = precision_recall_fscore_support(y_val, val_pred, average=None, zero_division=0)
            neutral_f1 = f1_per_class[1] if len(f1_per_class) > 1 else 0
            cv_acc.append(acc)
            cv_prec.append(prec)
            cv_rec.append(rec)
            cv_f1.append(f1)
            cv_neutral_f1.append(neutral_f1)
            print(f"  Fold acc: {acc:.4f}, prec: {prec:.4f}, rec: {rec:.4f}, f1: {f1:.4f}, neutral_f1: {neutral_f1:.4f}")
        print("\n=== 交差検証結果 ===")
        print(f"Accuracy: {np.mean(cv_acc):.4f} ± {np.std(cv_acc):.4f}")
        print(f"Precision: {np.mean(cv_prec):.4f} ± {np.std(cv_prec):.4f}")
        print(f"Recall: {np.mean(cv_rec):.4f} ± {np.std(cv_rec):.4f}")
        print(f"F1: {np.mean(cv_f1):.4f} ± {np.std(cv_f1):.4f}")
        print(f"Neutral F1: {np.mean(cv_neutral_f1):.4f} ± {np.std(cv_neutral_f1):.4f}")
        sample_weights = self.compute_sample_weights(y)
        final_model = xgb.XGBClassifier(**best_params)
        final_model.fit(X, y, sample_weight=sample_weights)
        preds = np.argmax(final_model.predict_proba(X_test), axis=1)
        print("\n=== テスト結果 (ホールドアウトセット) ===")
        print(classification_report(
            y_test, preds,
            target_names=['ネガティブ', '中立', 'ポジティブ'],
            digits=4
        ))
        acc = accuracy_score(y_test, preds)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test, preds, average='weighted')
        print(f"\nWeighted Metrics - Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")
        prec_per_class, rec_per_class, f1_per_class, support = precision_recall_fscore_support(y_test, preds, average=None, zero_division=0)
        print("\n各クラスごとの指標:")
        for i, class_name in enumerate(['ネガティブ', '中立', 'ポジティブ']):
            print(f"{class_name}: Precision={prec_per_class[i]:.4f}, Recall={rec_per_class[i]:.4f}, F1={f1_per_class[i]:.4f}, Support={support[i]}")
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(y_test, preds)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['ネガティブ', '中立', 'ポジティブ'])
        disp.plot(cmap="Blues")
        plt.title("Confusion Matrix (Test Set)")
        plt.show()
        importances = final_model.feature_importances_
        importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances}).sort_values(by="importance", ascending=False)
        print(f"\n=== 上位15の重要特徴量 ===")
        print(importance_df.head(15).to_string(index=False))
        plt.figure(figsize=(12, 8))
        top_features = importance_df.head(20)
        plt.barh(range(len(top_features)), top_features['importance'])
        plt.yticks(range(len(top_features)), top_features['feature'])
        plt.xlabel('Feature Importance')
        plt.title('Top 20 Feature Importances (XGBoost)')
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()
        try:
            print("\n=== SHAP解析 ===")
            explainer = shap.TreeExplainer(final_model)
            shap_values = explainer.shap_values(X_test)
            plt.figure(figsize=(12, 8))
            shap.summary_plot(shap_values, X_test, feature_names=feature_names, show=False, max_display=20)
            plt.title("SHAP Feature Importance Summary")
            plt.tight_layout()
            plt.show()
            shap_importance = np.mean(np.abs(shap_values), axis=0)
            shap_importance_df = pd.DataFrame({'feature': feature_names, 'shap_importance': shap_importance}).sort_values(by="shap_importance", ascending=False)
            print("SHAP重要度上位15特徴量:")
            print(shap_importance_df.head(15).to_string(index=False))
        except Exception as e:
            print(f"SHAP解析エラー: {e}")
            shap_values = None
            shap_importance_df = None
        test_df = test_df.copy()
        test_df["predicted_sentiment"] = preds
        return {
            "model": final_model,
            "label_encoder": self.label_encoder,
            "scaler": self.scaler,
            "test_df": test_df,
            "train_df": train_df,
            "feature_importance": importance_df,
            "shap_values": shap_values,
            "shap_importance": shap_importance_df if 'shap_importance_df' in locals() else None,
            "feature_names": feature_names,
            "cv_metrics": {
                "accuracy": cv_acc,
                "precision": cv_prec,
                "recall": cv_rec,
                "f1": cv_f1,
                "neutral_f1": cv_neutral_f1
            }
        }

if __name__ == "__main__":
    pipeline = JapaneseABSAXGBoostTabularPipeline()
    results = pipeline.fit()
```

```python
# Japanese ABSA XGBoost Tabular Pipeline: Fixed Version

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings, re, optuna, os, requests
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, precision_recall_fscore_support, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import xgboost as xgb
import shap

import matplotlib.pyplot as plt
import matplotlib
import os

# Download Noto Sans JP font if not exists (Colab/Kaggle)
font_url = "https://github.com/googlefonts/noto-cjk/raw/main/Sans/OTF/Japanese/NotoSansCJKjp-Regular.otf"
font_path = "/usr/share/fonts/truetype/noto/NotoSansCJKjp-Regular.otf"

if not os.path.exists(font_path):
    os.makedirs(os.path.dirname(font_path), exist_ok=True)
    import requests
    r = requests.get(font_url)
    with open(font_path, "wb") as f:
        f.write(r.content)

# Register and set as default
matplotlib.font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Noto Sans CJK JP'
plt.rcParams['axes.unicode_minus'] = False  # For minus sign

# Check (optional)
# print([f.name for f in matplotlib.font_manager.fontManager.ttflist if 'Noto' in f.name])

# Japanese language processing
try:
    import MeCab
    # Try different MeCab configurations
    try:
        import ipadic
        mecab = MeCab.Tagger(f'-d {ipadic.DICDIR} -Owakati')
        print("✅ MeCab loaded with ipadic dictionary.")
    except:
        # Fallback configurations
        try:
            mecab = MeCab.Tagger('-Owakati')
            print("✅ MeCab loaded with default dictionary.")
        except:
            mecab = MeCab.Tagger('')
            print("✅ MeCab loaded with basic configuration.")
except Exception as e:
    print("❌ MeCab failed:", e)
    mecab = None

warnings.filterwarnings('ignore')

class JapaneseBusinessInsightExtractor:
    def __init__(self):
        self.aspects = {
            'fit': ['フィット', 'サイズ', 'きつい', 'ゆるい', '小さい', '大きい', 'ちっちゃい', 'でかい', '合う', '合わない', 'ぴったり', 'ぶかぶか'],
            'quality': ['品質', '質', '素材', 'クオリティ', '安っぽい', '高級', '作り', '丈夫', 'しっかり', 'ちゃちい', 'ペラペラ', '生地'],
            'style': ['スタイル', 'デザイン', '色', 'おしゃれ', 'かわいい', 'かっこいい', 'ダサい', 'イケてる', 'オシャレ', '可愛い'],
            'comfort': ['快適', '心地', '着心地', '楽', 'らく', 'きもちいい', '気持ちいい', 'ふわふわ', 'さらさら', 'ごわごわ'],
            'price': ['価格', '値段', '金額', '高い', '安い', 'コスパ', 'お得', 'もったいない', 'ぼったくり', '激安', 'プチプラ'],
            'shipping': ['配送', '発送', '届く', '配達', '到着', '早い', '遅い', '梱包', 'はやい', 'おそい', 'すぐ']
        }
        self.positive_words = ['最高', '素晴らしい', '良い', 'いい', '完璧', '最強', 'やばい', 'すごい', 'めっちゃいい', 'マジいい', 'ヤバい', 'エモい', 'しあわせ', '幸せ']
        self.negative_words = ['最悪', 'ひどい', '悪い', 'だめ', 'クソ', 'うざい', 'ムカつく', 'イライラ', 'がっかり', 'ショック', 'つまらない', 'びみょう', '微妙']
        self.neutral_words = ['普通', 'まあまあ', '平均', 'ふつう', 'そこそこ', 'まずまず', '標準', 'ありきたり', 'こんなもん']

    def generate_business_recommendations(self, df):
        print("\n=== ビジネス分析と推奨事項 ===")
        for aspect in self.aspects.keys():
            aspect_col = f'aspect_{aspect}'
            if aspect_col in df.columns:
                aspect_mentions = df[aspect_col].sum()
                negative_mentions = df[(df[aspect_col] == 1) & (df['sentiment'] == 0)].shape[0]
                neutral_mentions = df[(df[aspect_col] == 1) & (df['sentiment'] == 1)].shape[0]
                if aspect_mentions > 0:
                    negative_rate = negative_mentions / aspect_mentions * 100
                    neutral_rate = neutral_mentions / aspect_mentions * 100
                    print(f"• {aspect.upper()}: {aspect_mentions} 件の言及, {negative_rate:.1f}% ネガティブ, {neutral_rate:.1f}% 中立")
        return {"status": "completed"}

def executive_summary(df):
    print("\n=== エグゼクティブサマリー ===")
    total_reviews = len(df)
    sentiment_dist = df['sentiment'].value_counts(normalize=True) * 100
    sentiment_labels = {0: 'ネガティブ', 1: '中立', 2: 'ポジティブ'}
    print(f"分析したレビュー総数: {total_reviews:,}")
    print("感情分布:")
    for sentiment, pct in sentiment_dist.items():
        print(f"  • {sentiment_labels.get(sentiment, sentiment)}: {pct:.1f}%")

def plot_overall_sentiment(df):
    plt.figure(figsize=(15, 5))
    plt.rcParams['font.family'] = 'DejaVu Sans'  # For Japanese text display
    
    plt.subplot(1, 3, 1)
    sentiment_counts = df['sentiment'].value_counts()
    sentiment_labels = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    labels = [sentiment_labels.get(k, str(k)) for k in sentiment_counts.index]
    colors = ['#ff6b6b', '#ffd93d', '#6bcf7f']
    plt.pie(sentiment_counts.values, labels=labels, autopct='%1.1f%%',
            colors=colors, startangle=90)
    plt.title('Sentiment Distribution')
    
    plt.subplot(1, 3, 2)
    plt.hist(df['sentiment'], bins=3, color='skyblue', alpha=0.7, edgecolor='black')
    plt.xlabel('Sentiment')
    plt.ylabel('Frequency')
    plt.title('Sentiment Distribution')
    plt.xticks([0, 1, 2], ['Negative', 'Neutral', 'Positive'])
    
    plt.subplot(1, 3, 3)
    sentiment_counts_df = df['sentiment'].value_counts().reindex([0, 1, 2], fill_value=0)
    plt.bar(range(3), sentiment_counts_df.values, color=colors, alpha=0.7)
    plt.title('Sentiment Counts')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.xticks(range(3), ['Negative', 'Neutral', 'Positive'])
    
    plt.tight_layout()
    plt.show()

def plot_negative_aspect_breakdown(df):
    aspects = ['fit', 'quality', 'style', 'comfort', 'price', 'shipping']
    negative_counts, neutral_counts = [], []
    
    for aspect in aspects:
        aspect_col = f'aspect_{aspect}'
        if aspect_col in df.columns:
            neg_count = df[(df[aspect_col] == 1) & (df['sentiment'] == 0)].shape[0]
            neu_count = df[(df[aspect_col] == 1) & (df['sentiment'] == 1)].shape[0]
            negative_counts.append(neg_count)
            neutral_counts.append(neu_count)
        else:
            negative_counts.append(0)
            neutral_counts.append(0)
    
    plt.figure(figsize=(12, 6))
    x = np.arange(len(aspects))
    width = 0.35
    
    plt.bar(x - width/2, negative_counts, width, label='Negative', color='#ff6b6b', alpha=0.7)
    plt.bar(x + width/2, neutral_counts, width, label='Neutral', color='#ffd93d', alpha=0.7)
    
    plt.title('Negative and Neutral Sentiment by Aspect')
    plt.xlabel('Aspects')
    plt.ylabel('Number of Reviews')
    plt.xticks(x, aspects, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_aspect_sentiment_heatmap(df):
    aspects = ['fit', 'quality', 'style', 'comfort', 'price', 'shipping']
    sentiments = [0, 1, 2]  # negative, neutral, positive
    sentiment_labels = ['Negative', 'Neutral', 'Positive']
    
    heatmap_data = []
    for aspect in aspects:
        aspect_col = f'aspect_{aspect}'
        row = []
        for sentiment in sentiments:
            if aspect_col in df.columns:
                count = df[(df[aspect_col] == 1) & (df['sentiment'] == sentiment)].shape[0]
                row.append(count)
            else:
                row.append(0)
        heatmap_data.append(row)
    
    plt.figure(figsize=(10, 6))
    sns.heatmap(heatmap_data,
                xticklabels=sentiment_labels,
                yticklabels=aspects,
                annot=True,
                fmt='d',
                cmap='RdYlBu_r',
                cbar_kws={'label': 'Number of Reviews'})
    plt.title('Aspect-Sentiment Heatmap')
    plt.xlabel('Sentiment')
    plt.ylabel('Aspects')
    plt.tight_layout()
    plt.show()

class JapaneseABSAXGBoostTabularPipeline:
    def __init__(self, max_train_samples=8000, max_test_samples=2000, n_splits=5, tfidf_max_features=200):
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.insight_extractor = JapaneseBusinessInsightExtractor()
        self.max_train_samples = max_train_samples
        self.max_test_samples = max_test_samples
        self.n_splits = n_splits
        self.feature_cols = []
        self.tfidf_max_features = tfidf_max_features
        self.tfidf_vec = TfidfVectorizer(max_features=tfidf_max_features, token_pattern=r'\b\w+\b')
        
    def tokenize_japanese(self, text):
        """Simple Japanese tokenization using MeCab if available, else char-level fallback."""
        if mecab is not None:
            try:
                result = mecab.parse(text).strip()
                # Remove empty strings and newlines
                tokens = [token for token in result.split() if token.strip()]
                return ' '.join(tokens) if tokens else text
            except Exception as e:
                print(f"MeCab parse error: {e}")
                return ' '.join(list(text))
        return ' '.join(list(text))

    def load_and_preprocess_data(self):
        print("WRIME データセットを読み込み中...")
        wrime_url = "https://raw.githubusercontent.com/ids-cv/wrime/refs/heads/master/wrime-ver1.tsv"
        wrime_path = "/kaggle/working/wrime-ver1.tsv"
        
        try:
            if not os.path.exists(wrime_path):
                print("WRIME データセットをダウンロード中...")
                r = requests.get(wrime_url)
                r.raise_for_status()
                with open(wrime_path, "wb") as f:
                    f.write(r.content)
            
            df_wrime = pd.read_csv(wrime_path, sep="\t")
            df_wrime = df_wrime.dropna(subset=["Sentence"])
            
            def get_sentiment(row):
                joy = row.get('Avg. Readers_Joy', 0)
                sadness = row.get('Avg. Readers_Sadness', 0)
                anger = row.get('Avg. Readers_Anger', 0)
                
                if joy > max(sadness, anger):
                    return 2  # positive
                elif (sadness > joy) or (anger > joy):
                    return 0  # negative
                else:
                    return 1  # neutral
            
            df_wrime['sentiment'] = df_wrime.apply(get_sentiment, axis=1)
            df_wrime = df_wrime.rename(columns={'Sentence': 'review_text'})
            df_wrime['rating'] = df_wrime['sentiment'] + 1
            
            print(f"WRIME データセットから {len(df_wrime)} 件のレビューを読み込みました")
            
        except Exception as e:
            print(f"データセット読み込みエラー: {e}")
            print("合成日本語データを使用します...")
            return self._create_synthetic_japanese_data()
        
        train_df, test_df = train_test_split(df_wrime, test_size=0.2, random_state=42, stratify=df_wrime['sentiment'])
        train_df = train_df.head(self.max_train_samples)
        test_df = test_df.head(self.max_test_samples)
        
        return train_df, test_df

    def _create_synthetic_japanese_data(self):
        print("合成日本語レビューデータを作成中...")
        np.random.seed(42)
        
        pos_templates = [
            "この{item}は{aspect}が{positive_word}です！",
            "{item}の{aspect}がマジで{positive_word}！",
            "めっちゃ{positive_word}な{item}、{aspect}も最高",
            "{positive_word}すぎる{item}！{aspect}がヤバい",
            "この{item}、{aspect}が{positive_word}でおすすめ"
        ]
        
        neg_templates = [
            "この{item}は{aspect}が{negative_word}です",
            "{item}の{aspect}がクソ{negative_word}",
            "マジで{negative_word}な{item}、{aspect}も最悪",
            "{negative_word}すぎる{item}…{aspect}がダメ",
            "この{item}、{aspect}が{negative_word}でがっかり"
        ]
        
        neu_templates = [
            "この{item}は{aspect}が普通です",
            "{item}の{aspect}はまあまあかな",
            "普通の{item}、{aspect}は可もなく不可もなく",
            "この{item}、{aspect}はそこそこ",
            "{item}の{aspect}は標準的"
        ]
        
        items = ["服", "トップス", "ブラウス", "セーター", "ジャケット", "パンツ", "スカート", "ワンピース"]
        aspects = ["フィット感", "品質", "デザイン", "着心地", "素材", "色"]
        pos_words = ["良い", "素晴らしい", "最高", "完璧", "いい感じ", "エモい"]
        neg_words = ["悪い", "ひどい", "最悪", "ダメ", "微妙", "イマイチ"]
        
        n_samples = self.max_train_samples + self.max_test_samples
        reviews, sentiments = [], []
        
        for i in range(n_samples):
            rand = np.random.random()
            if rand < 0.6:  # positive
                review = np.random.choice(pos_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects),
                    positive_word=np.random.choice(pos_words)
                )
                sentiment = 2
            elif rand < 0.8:  # neutral
                review = np.random.choice(neu_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects)
                )
                sentiment = 1
            else:  # negative
                review = np.random.choice(neg_templates).format(
                    item=np.random.choice(items),
                    aspect=np.random.choice(aspects),
                    negative_word=np.random.choice(neg_words)
                )
                sentiment = 0
            
            reviews.append(review)
            sentiments.append(sentiment)
        
        df = pd.DataFrame({'review_text': reviews, 'sentiment': sentiments})
        df['rating'] = df['sentiment'] + 1
        
        train_df = df.iloc[:self.max_train_samples].copy()
        test_df = df.iloc[self.max_train_samples:self.max_train_samples + self.max_test_samples].copy()
        
        return train_df, test_df

    def extract_comprehensive_features(self, df, fit_tfidf=False):
        print("包括的特徴量とTF-IDFを抽出中...")
        
        # Make a copy to avoid modifying original
        df = df.copy()
        
        # Aspect detection
        aspects = self.insight_extractor.aspects
        for aspect, keywords in aspects.items():
            pattern = '|'.join([re.escape(kw) for kw in keywords])
            df[f'aspect_{aspect}'] = df['review_text'].str.contains(pattern, na=False, regex=True).astype(int)
        
        # Basic text features
        df['text_length'] = df['review_text'].str.len()
        df['char_count'] = df['review_text'].str.len()
        df['hiragana_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[あ-ん]', x)))
        df['katakana_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[ア-ン]', x)))
        df['kanji_count'] = df['review_text'].apply(lambda x: len(re.findall(r'[一-龯]', x)))
        
        # Sentence structure features
        df['sentence_count'] = df['review_text'].str.count(r'[。！？]') + 1
        df['exclamation_count'] = df['review_text'].str.count('！') + df['review_text'].str.count('!')
        
        # Fixed the regex issue: escape the question mark
        df['question_count'] = df['review_text'].str.count(r'\？') + df['review_text'].str.count(r'\?')
        
        df['small_tsu_count'] = df['review_text'].str.count('っ')
        df['elongated_vowel_count'] = df['review_text'].str.count('ー')
        
        # Sentiment word features
        positive_words = self.insight_extractor.positive_words
        negative_words = self.insight_extractor.negative_words
        neutral_words = self.insight_extractor.neutral_words
        
        def count_sentiment_words(text, word_list):
            return sum(1 for word in word_list if word in text)
        
        df['positive_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, positive_words))
        df['negative_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, negative_words))
        df['neutral_word_count'] = df['review_text'].apply(lambda x: count_sentiment_words(x, neutral_words))
        
        # Interaction features
        for aspect in aspects.keys():
            aspect_col = f'aspect_{aspect}'
            if aspect_col in df.columns:
                df[f'{aspect}_positive_interaction'] = df[aspect_col] * df['positive_word_count']
                df[f'{aspect}_negative_interaction'] = df[aspect_col] * df['negative_word_count']
        
        # Additional features
        df['balanced_sentiment'] = ((df['positive_word_count'] > 0) & (df['negative_word_count'] > 0)).astype(int)
        df['short_review'] = (df['text_length'] <= 10).astype(int)
        df['neutral_indicators'] = df['neutral_word_count']
        
        # TF-IDF Features with Japanese tokenization
        try:
            if fit_tfidf:
                tokenized_texts = df['review_text'].apply(self.tokenize_japanese)
                tfidf_matrix = self.tfidf_vec.fit_transform(tokenized_texts)
            else:
                tokenized_texts = df['review_text'].apply(self.tokenize_japanese)
                tfidf_matrix = self.tfidf_vec.transform(tokenized_texts)
            
            tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), 
                                   columns=[f"tfidf_{i}" for i in range(self.tfidf_max_features)], 
                                   index=df.index)
            df = pd.concat([df, tfidf_df], axis=1)
        except Exception as e:
            print(f"TF-IDF extraction error: {e}")
            # Add dummy TF-IDF columns if extraction fails
            for i in range(self.tfidf_max_features):
                df[f"tfidf_{i}"] = 0.0
        
        # Define feature columns (excluding TF-IDF and embeddings for importance analysis)
        exclude_cols = ['review_text', 'sentiment', 'rating']
        potential_features = [
            col for col in df.columns
            if col not in exclude_cols and
               not col.startswith("tfidf_") and
               not col.startswith("emb_") and
               not col.startswith("Avg.") and
               not col.startswith("high_")
        ]
        
        numeric_features = [col for col in potential_features if pd.api.types.is_numeric_dtype(df[col])]
        self.feature_cols = numeric_features
        
        return df

    def compute_sample_weights(self, y):
        from sklearn.utils.class_weight import compute_class_weight
        classes = np.unique(y)
        class_weights = compute_class_weight('balanced', classes=classes, y=y)
        
        # Boost neutral class weight
        neutral_class_idx = np.where(classes == 1)[0]
        if len(neutral_class_idx) > 0:
            class_weights[neutral_class_idx[0]] *= 1.5
        
        weights = np.zeros_like(y, dtype=float)
        for idx, cls in enumerate(classes):
            weights[y == cls] = class_weights[idx]
        
        return weights

    def fit(self):
        print("日本語ABSA XGBoost (表形式特徴量のみ) パイプラインを開始...")
        
        # Load and preprocess data
        train_df, test_df = self.load_and_preprocess_data()
        train_df = self.extract_comprehensive_features(train_df, fit_tfidf=True)
        test_df = self.extract_comprehensive_features(test_df, fit_tfidf=False)
        
        # Generate insights and visualizations
        executive_summary(train_df)
        plot_overall_sentiment(train_df)
        plot_negative_aspect_breakdown(train_df)
        plot_aspect_sentiment_heatmap(train_df)
        self.insight_extractor.generate_business_recommendations(train_df)
        
        # Prepare features
        y = train_df['sentiment'].values
        y_test = test_df['sentiment'].values
        X = train_df[self.feature_cols].values
        X_test = test_df[self.feature_cols].values
        
        # Scale features
        X = self.scaler.fit_transform(X)
        X_test = self.scaler.transform(X_test)
        
        feature_names = self.feature_cols
        
        # Cross-validation setup
        skf = StratifiedKFold(n_splits=self.n_splits, shuffle=True, random_state=42)
        train_idx, val_idx = next(skf.split(X, y))
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        weights_train = self.compute_sample_weights(y_train)
        
        # Hyperparameter optimization
        def objective(trial):
            param = {
                "objective": "multi:softprob",
                "num_class": 3,
                "eval_metric": "mlogloss",
                "max_depth": trial.suggest_int("max_depth", 3, 8),
                "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
                "n_estimators": trial.suggest_int("n_estimators", 100, 500),
                "subsample": trial.suggest_float("subsample", 0.7, 1.0),
                "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0),
                "gamma": trial.suggest_float("gamma", 0, 5),
                "lambda": trial.suggest_float("lambda", 1e-3, 10.0, log=True),
                "alpha": trial.suggest_float("alpha", 1e-3, 10.0, log=True),
                "min_child_weight": trial.suggest_int("min_child_weight", 1, 7),
                "tree_method": "auto",
                "use_label_encoder": False,
                "verbosity": 0,
                "random_state": 42
            }
            
            model = xgb.XGBClassifier(**param)
            model.fit(X_train, y_train, sample_weight=weights_train)
            preds = np.argmax(model.predict_proba(X_val), axis=1)
            
            _, _, f1, _ = precision_recall_fscore_support(y_val, preds, average='weighted', zero_division=0)
            return 1.0 - f1
        
        print("Optuna チューニング (重み付きF1スコアを最適化)...")
        study = optuna.create_study(direction="minimize")
        study.optimize(objective, n_trials=20, show_progress_bar=False)
        
        best_params = study.best_trial.params
        best_params.update({
            "objective": "multi:softprob",
            "num_class": 3,
            "use_label_encoder": False,
            "tree_method": "auto",
            "random_state": 42
        })
        print("最適パラメータ:", best_params)
        
        # Cross-validation
        oof_preds = np.zeros((len(y), 3))
        cv_acc, cv_prec, cv_rec, cv_f1 = [], [], [], []
        cv_neutral_f1 = []
        
        for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
            print(f"Fold {fold+1}/{self.n_splits}...")
            X_tr, X_val = X[train_idx], X[val_idx]
            y_tr, y_val = y[train_idx], y[val_idx]
            
            weights = self.compute_sample_weights(y_tr)
            model = xgb.XGBClassifier(**best_params)
            model.fit(X_tr, y_tr, sample_weight=weights)
            
            val_pred = np.argmax(model.predict_proba(X_val), axis=1)
            oof_preds[val_idx] = model.predict_proba(X_val)
            
            acc = accuracy_score(y_val, val_pred)
            prec, rec, f1, _ = precision_recall_fscore_support(y_val, val_pred, average='weighted', zero_division=0)
            prec_per_class, rec_per_class, f1_per_class, _ = precision_recall_fscore_support(y_val, val_pred, average=None, zero_division=0)
            
            neutral_f1 = f1_per_class[1] if len(f1_per_class) > 1 else 0
            
            cv_acc.append(acc)
            cv_prec.append(prec)
            cv_rec.append(rec)
            cv_f1.append(f1)
            cv_neutral_f1.append(neutral_f1)
            
            print(f"  Fold acc: {acc:.4f}, prec: {prec:.4f}, rec: {rec:.4f}, f1: {f1:.4f}, neutral_f1: {neutral_f1:.4f}")
        
        print("\n=== 交差検証結果 ===")
        print(f"Accuracy: {np.mean(cv_acc):.4f} ± {np.std(cv_acc):.4f}")
        print(f"Precision: {np.mean(cv_prec):.4f} ± {np.std(cv_prec):.4f}")
        print(f"Recall: {np.mean(cv_rec):.4f} ± {np.std(cv_rec):.4f}")
        print(f"F1: {np.mean(cv_f1):.4f} ± {np.std(cv_f1):.4f}")
        print(f"Neutral F1: {np.mean(cv_neutral_f1):.4f} ± {np.std(cv_neutral_f1):.4f}")
        
                # Final model training
        sample_weights = self.compute_sample_weights(y)
        final_model = xgb.XGBClassifier(**best_params)
        final_model.fit(X, y, sample_weight=sample_weights)

        # Test prediction and evaluation
        preds = np.argmax(final_model.predict_proba(X_test), axis=1)
        print("\n=== テスト結果 (ホールドアウトセット) ===")
        print(classification_report(
            y_test, preds,
            target_names=['ネガティブ', '中立', 'ポジティブ'],
            digits=4
        ))
        acc = accuracy_score(y_test, preds)
        prec, rec, f1, _ = precision_recall_fscore_support(y_test, preds, average='weighted')
        print(f"\nWeighted Metrics - Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

        # Per-class Metrics
        prec_per_class, rec_per_class, f1_per_class, support = precision_recall_fscore_support(
            y_test, preds, average=None, zero_division=0)
        print("\n各クラスごとの指標:")
        for i, class_name in enumerate(['ネガティブ', '中立', 'ポジティブ']):
            print(f"{class_name}: Precision={prec_per_class[i]:.4f}, Recall={rec_per_class[i]:.4f}, F1={f1_per_class[i]:.4f}, Support={support[i]}")

        # Confusion Matrix
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(y_test, preds)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['ネガティブ', '中立', 'ポジティブ'])
        disp.plot(cmap="Blues")
        plt.title("Confusion Matrix (Test Set)")
        plt.show()

        # Feature Importance
        importances = final_model.feature_importances_
        importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances}).sort_values(by="importance", ascending=False)
        print(f"\n=== 上位15の重要特徴量 ===")
        print(importance_df.head(15).to_string(index=False))
        plt.figure(figsize=(12, 8))
        top_features = importance_df.head(20)
        plt.barh(range(len(top_features)), top_features['importance'])
        plt.yticks(range(len(top_features)), top_features['feature'])
        plt.xlabel('Feature Importance')
        plt.title('Top 20 Feature Importances (XGBoost)')
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()

        # SHAP Analysis
        try:
            print("\n=== SHAP解析 ===")
            explainer = shap.TreeExplainer(final_model)
            shap_values = explainer.shap_values(X_test)
            plt.figure(figsize=(12, 8))
            shap.summary_plot(shap_values, X_test, feature_names=feature_names, show=False, max_display=20)
            plt.title("SHAP Feature Importance Summary")
            plt.tight_layout()
            plt.show()
            shap_importance = np.mean(np.abs(shap_values), axis=0)
            shap_importance_df = pd.DataFrame({'feature': feature_names, 'shap_importance': shap_importance}).sort_values(by="shap_importance", ascending=False)
            print("SHAP重要度上位15特徴量:")
            print(shap_importance_df.head(15).to_string(index=False))
        except Exception as e:
            print(f"SHAP解析エラー: {e}")
            shap_values = None
            shap_importance_df = None

        # Attach prediction to test_df
        test_df = test_df.copy()
        test_df["predicted_sentiment"] = preds

        return {
            "model": final_model,
            "label_encoder": self.label_encoder,
            "scaler": self.scaler,
            "test_df": test_df,
            "train_df": train_df,
            "feature_importance": importance_df,
            "shap_values": shap_values,
            "shap_importance": shap_importance_df if 'shap_importance_df' in locals() else None,
            "feature_names": feature_names,
            "cv_metrics": {
                "accuracy": cv_acc,
                "precision": cv_prec,
                "recall": cv_rec,
                "f1": cv_f1,
                "neutral_f1": cv_neutral_f1
            }
        }

# Run the pipeline
if __name__ == "__main__":
    pipeline = JapaneseABSAXGBoostTabularPipeline()
    results = pipeline.fit()
```
