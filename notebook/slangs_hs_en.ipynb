```python
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"

import pandas as pd
import numpy as np
import requests
import io
import json
import re

# ========================
# 1. Load All Datasets
# ========================
# --- 1.1 Sarcasm Headlines ---
url_sarcasm = "https://raw.githubusercontent.com/rishabhmisra/News-Headlines-Dataset-For-Sarcasm-Detection/master/Sarcasm_Headlines_Dataset.json"
sarcasm_data = [json.loads(line) for line in requests.get(url_sarcasm).text.strip().split('\n')]
df_sarcasm = pd.DataFrame(sarcasm_data)[['headline', 'is_sarcastic']]
df_sarcasm = df_sarcasm.rename(columns={'headline':'text', 'is_sarcastic':'label'})
df_sarcasm['task'] = 'sarcasm'

# --- 1.2 Hate Speech Dataset ---
url_hate = "https://raw.githubusercontent.com/t-davidson/hate-speech-and-offensive-language/master/data/labeled_data.csv"
df_hate = pd.read_csv(url_hate)
df_hate = df_hate.rename(columns={'tweet':'text', 'class':'label'})
df_hate['task'] = 'hate_speech'

# --- 1.3 Gen-Z Slang Dataset (from HuggingFace) ---
from datasets import load_dataset
ds_genz = load_dataset("MLBtrio/genz-slang-dataset", split="train")
df_genz = pd.DataFrame(ds_genz)[['Slang', 'Description', 'Example', 'Context']]
df_genz['text'] = df_genz['Slang'] + " : " + df_genz['Description']
df_genz['label'] = 2  # Neutral, as no sentiment label
df_genz['task'] = 'slang'
df_genz = df_genz[['text', 'label', 'task']]

# ========================
# 2. Slang & Emoji Dictionary with Fallbacks
# ========================
# 2.1 Load Gen-Z Slang CSV
url_slang = "https://raw.githubusercontent.com/kaspercools/genz-dataset/refs/heads/main/genz_slang.csv"
resp_slang = requests.get(url_slang)
df_slang = pd.read_csv(io.StringIO(resp_slang.text))
slang_map = {
    str(row['keyword']).strip().lower(): str(row['description']).strip()
    for _, row in df_slang.iterrows()
    if pd.notnull(row['keyword']) and pd.notnull(row['description'])
}

# 2.2 Load Gen-Z Emojis CSV
url_emoji = "https://raw.githubusercontent.com/kaspercools/genz-dataset/refs/heads/main/genz_emojis.csv"
resp_emoji = requests.get(url_emoji)
df_emoji = pd.read_csv(io.StringIO(resp_emoji.text))
emoji_map = {
    str(row['emoji']).strip(): str(row['Description']).strip()
    for _, row in df_emoji.iterrows()
    if pd.notnull(row['emoji']) and pd.notnull(row['Description'])
}

# 2.3 Phrase Variants and Manual Backups
variant_patterns = {
    "fleek": ["on fleek"],
    "cap": ["no cap"],
    "shade": ["throw shade"],
    "tea": ["spill the tea"],
    "key": ["low key", "high key"],
    "bestie": ["bestie vibes"],
    "grass": ["touch grass"]
}
custom_phrase_map = {}
for base, phrases in variant_patterns.items():
    if base in slang_map:
        for phrase in phrases:
            custom_phrase_map[phrase] = slang_map[base]

manual_phrase_map = {
    "on fleek": "something that is perfect or done really well",
}

# Merge: manual > custom > slang > emoji
translation_map = {**manual_phrase_map, **custom_phrase_map, **slang_map, **emoji_map}

# --- Proof that the mapping works
print("'on fleek' in translation_map?", "on fleek" in translation_map)
print("Value for 'on fleek':", translation_map.get("on fleek"))
print("Keys containing 'fleek':", [k for k in translation_map if 'fleek' in k])

# ========================
# 2.4 Sentiment Tagging for Slang/Emoji (optional, for your review)
positive_words = [
    "good", "amazing", "great", "excellent", "love", "awesome", "hype",
    "exciting", "famous", "best", "cool", "excited", "fun", "hilarious", "positive", "win", "happy"
]
negative_words = [
    "bad", "awkward", "embarrassing", "cringe", "sad", "hate", "annoying",
    "upset", "bitter", "negative", "lame", "worst", "angry", "problem", "fail", "boring", "entitled"
]
neutral_words = [
    "person", "thing", "expression", "gesture", "object", "action", "event", "saying"
]

def tag_sentiment(definition):
    definition_lower = definition.lower()
    if any(word in definition_lower for word in positive_words):
        return "positive"
    elif any(word in definition_lower for word in negative_words):
        return "negative"
    elif any(word in definition_lower for word in neutral_words):
        return "neutral"
    else:
        return "ambiguous"

sentiment_df = pd.DataFrame([
    {"token": k, "definition": v, "sentiment": tag_sentiment(v)}
    for k, v in translation_map.items()
])

print("\nAmbiguous or Neutral translations for manual review:")
print(sentiment_df[sentiment_df['sentiment'].isin(['ambiguous', 'neutral'])][['token', 'definition']].head(8))

# ========================
# 2.5 Replacement function (with verbose mapping output)
def replace_slang_and_emoji(text, translation_map, verbose=False):
    if pd.isna(text):
        return ""
    text = str(text)
    mapping_used = []
    for key in sorted(translation_map.keys(), key=lambda x: -len(x)):
        val = translation_map[key]
        if re.match(r'^\W+$', key):
            if key in text and verbose:
                mapping_used.append((key, val))
            text = text.replace(key, val)
        else:
            pattern = r'(?i)(?<!\w){}(?=\W|$)'.format(re.escape(key))
            if re.search(pattern, text):
                if verbose:
                    mapping_used.append((key, val))
                text = re.sub(pattern, val, text)
    if verbose:
        print("Mappings used in this sentence:")
        if mapping_used:
            for k, v in mapping_used:
                print(f"  '{k}' => '{v}'")
        else:
            print("  (None)")
    return text

# --- Demo
demo_sentence = "I'm dead 😂, this party is on fleek!"
print("\n--- Demo Replacement ---")
print("Original:", demo_sentence)
translated = replace_slang_and_emoji(demo_sentence, translation_map, verbose=True)
print("Translated:", translated)

# ========================
# 3. Translate Sarcasm & Hate Datasets
# ========================
print("Translating slang and emojis in datasets...")
df_sarcasm['text'] = df_sarcasm['text'].apply(lambda t: replace_slang_and_emoji(t, translation_map))
df_hate['text'] = df_hate['text'].apply(lambda t: replace_slang_and_emoji(t, translation_map))

# ========================
# 4. Combine All, Harmonize, Sample for Colab
# ========================
df_sarcasm['label'] = df_sarcasm['label'].astype(int)
df_hate['label'] = df_hate['label'].astype(int)
df_genz['label'] = df_genz['label'].astype(int)
df_sarcasm_copy = df_sarcasm.copy()
df_sarcasm_copy.loc[df_sarcasm_copy['label']==1, 'label'] = 0
df_sarcasm_copy.loc[df_sarcasm_copy['label']==0, 'label'] = 2

df_combined = pd.concat([
    df_sarcasm_copy[df_sarcasm_copy['label'].isin([0,2])],
    df_hate[df_hate['label'].isin([0,1,2])],
    df_genz
], ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)
print("Combined dataset shape:", df_combined.shape)
print("Label distribution:")
print(df_combined['label'].value_counts())

df_combined = df_combined.dropna(subset=['text', 'label'])
df_combined = df_combined[df_combined['text'].str.len() > 0]
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)
N_train, N_val, N_test = 10_000, 300, 300
total_needed = N_train + N_val + N_test

if len(df_combined) < total_needed:
    print(f"Warning: Only {len(df_combined)} samples available, adjusting sample sizes...")
    N_train = min(N_train, len(df_combined) - 600)
    N_val = min(N_val, (len(df_combined) - N_train) // 2)
    N_test = len(df_combined) - N_train - N_val

df_train = df_combined.iloc[:N_train]
df_valid = df_combined.iloc[N_train:N_train+N_val]
df_test  = df_combined.iloc[N_train+N_val:N_train+N_val+N_test]
print(f"Train: {len(df_train)}, Valid: {len(df_valid)}, Test: {len(df_test)}")
num_labels = len(df_combined['label'].unique())
print(f"Number of labels: {num_labels}")

# ========================
# 5. LLM Fine-Tuning (DistilBERT, HuggingFace Trainer)
# ========================
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import matplotlib.pyplot as plt

model_name = "bert-base-uncased"  # Can use other models if needed
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

def tokenize(batch):
    return tokenizer(batch['text'], truncation=True, padding='max_length', max_length=128)

import datasets
ds_train = datasets.Dataset.from_pandas(df_train.rename(columns={'label':'labels'}))
ds_valid = datasets.Dataset.from_pandas(df_valid.rename(columns={'label':'labels'}))
ds_test  = datasets.Dataset.from_pandas(df_test.rename(columns={'label':'labels'}))
print("Tokenizing datasets...")
ds_train = ds_train.map(tokenize, batched=True)
ds_valid = ds_valid.map(tokenize, batched=True)
ds_test = ds_test.map(tokenize, batched=True)
cols = ['input_ids', 'attention_mask', 'labels']
ds_train.set_format(type='torch', columns=cols)
ds_valid.set_format(type='torch', columns=cols)
ds_test.set_format(type='torch', columns=cols)

training_args = TrainingArguments(
    output_dir="./outputs",
    save_strategy="no",
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    logging_strategy="epoch",
    learning_rate=2e-5,
    weight_decay=0.01,
    report_to="none"
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=ds_train,
    eval_dataset=ds_valid,
    compute_metrics=None,
)
print("Starting model training...")
trainer.train()

# ========================
# 6. Evaluate Fine-tuned LLM
# ========================
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score, RocCurveDisplay

print("Evaluating fine-tuned model...")
preds = trainer.predict(ds_test)
y_true = df_test['label'].values
y_pred = np.argmax(preds.predictions, axis=1)
y_proba = torch.softmax(torch.tensor(preds.predictions), dim=1).numpy()

print("LLM (fine-tuned) classification report:")
print(classification_report(y_true, y_pred, digits=4))
cm = confusion_matrix(y_true, y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("LLM (Fine-tuned) Confusion Matrix")
plt.show()

try:
    if num_labels == 2:
        auc_roc = roc_auc_score(y_true, y_proba[:, 1])
        RocCurveDisplay.from_predictions(y_true, y_proba[:, 1])
    else:
        auc_roc = roc_auc_score(y_true, y_proba, multi_class='ovr')
        for i in range(num_labels):
            RocCurveDisplay.from_predictions((y_true == i).astype(int), y_proba[:, i], name=f'Class {i}')
    plt.title("LLM (Fine-tuned) AUC-ROC Curve")
    plt.show()
    print("AUC-ROC:", auc_roc)
except Exception as e:
    print(f"Could not calculate AUC-ROC: {e}")

# ========================
# 7. Embeddings from Fine-tuned LLM
# ========================
ft_model = trainer.model.base_model
ft_model.eval()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
ft_model.to(device)

def get_ft_embeddings(texts, tokenizer, model, batch_size=32, max_length=128):
    all_emb = []
    model.eval()
    for i in range(0, len(texts), batch_size):
        batch = texts.iloc[i:i+batch_size].tolist()
        tokens = tokenizer(batch, truncation=True, padding=True, max_length=max_length, return_tensors='pt')
        with torch.no_grad():
            input_ids = tokens['input_ids'].to(model.device)
            attention_mask = tokens['attention_mask'].to(model.device)
            outputs = model(input_ids, attention_mask=attention_mask)
            emb = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
            all_emb.append(emb)
    return np.vstack(all_emb)

print("Extracting embeddings...")
X_train_emb = get_ft_embeddings(df_train['text'], tokenizer, ft_model)
X_valid_emb = get_ft_embeddings(df_valid['text'], tokenizer, ft_model)
X_test_emb  = get_ft_embeddings(df_test['text'], tokenizer, ft_model)
print(f"Embedding shapes - Train: {X_train_emb.shape}, Valid: {X_valid_emb.shape}, Test: {X_test_emb.shape}")

# ========================
# 8. XGBoost + Optuna on LLM Embeddings
# ========================
import optuna
import xgboost as xgb
from sklearn.metrics import f1_score

def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 50, 150),
        "max_depth": trial.suggest_int("max_depth", 3, 8),
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, 0.3, log=True),
        "subsample": trial.suggest_float("subsample", 0.7, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0),
        "eval_metric": "mlogloss",
        "use_label_encoder": False,
        "random_state": 42,
        "tree_method": "hist",    # updated!
        "device": "cuda",         # updated!
    }
    clf = xgb.XGBClassifier(**params)
    clf.fit(X_train_emb, df_train['label'])
    preds = clf.predict(X_valid_emb)
    return f1_score(df_valid['label'], preds, average="weighted")

print("Optimizing XGBoost hyperparameters...")
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)
print("Best params:", study.best_params)

clf = xgb.XGBClassifier(**study.best_params, eval_metric='mlogloss', use_label_encoder=False, tree_method="hist", device="cuda")
clf.fit(X_train_emb, df_train['label'])
y_pred = clf.predict(X_test_emb)
y_proba = clf.predict_proba(X_test_emb)

# ========================
# 9. XGBoost Metrics & Visualization
# ========================
print("XGBoost (on FT LLM embeddings) classification report:")
print(classification_report(df_test['label'], y_pred, digits=4))
cm = confusion_matrix(df_test['label'], y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("XGBoost (on FT LLM embeddings) Confusion Matrix")
plt.show()

try:
    if num_labels == 2:
        auc_roc = roc_auc_score(df_test['label'], y_proba[:, 1])
        RocCurveDisplay.from_predictions(df_test['label'], y_proba[:, 1])
    else:
        auc_roc = roc_auc_score(df_test['label'], y_proba, multi_class='ovr')
        for i in range(num_labels):
            RocCurveDisplay.from_predictions((df_test['label'] == i).astype(int), y_proba[:, i], name=f'Class {i}')
    plt.title("XGBoost (on FT LLM) AUC-ROC Curve")
    plt.show()
    print("XGBoost AUC-ROC:", auc_roc)
except Exception as e:
    print(f"Could not calculate AUC-ROC for XGBoost: {e}")

# ========================
# 10. Feature Importance & SHAP (for completeness, not always meaningful)
# ========================
import shap
print("Generating SHAP explanations (embedding features)...")
try:
    shap_sample_size = min(100, len(X_test_emb))
    X_test_sample = X_test_emb[:shap_sample_size]
    explainer = shap.Explainer(clf, X_train_emb[:500])
    shap_values = explainer(X_test_sample)
    shap.summary_plot(shap_values, X_test_sample, 
                     feature_names=[f"emb_{i}" for i in range(X_test_sample.shape[1])],
                     show=False)
    plt.title("SHAP Summary Plot")
    plt.show()
except Exception as e:
    print(f"SHAP analysis failed: {e}")
try:
    xgb.plot_importance(clf, max_num_features=20)
    plt.title("XGBoost Feature Importance (LLM Embeddings)")
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Feature importance plot failed: {e}")

print("Analysis complete! Note: SHAP/feature importances above are for embedding dimensions, not real 'words'.")

# ========================
# 11. Demo Function for New Text
# ========================
def predict_sentiment(text, model=clf, tokenizer=tokenizer, ft_model=ft_model):
    print("\n--- Demo Replacement ---")
    print("Original:", text)
    translated_text = replace_slang_and_emoji(text, translation_map, verbose=True)
    print("Translated:", translated_text)
    embedding = get_ft_embeddings(pd.Series([translated_text]), tokenizer, ft_model)
    prediction = model.predict(embedding)[0]
    probabilities = model.predict_proba(embedding)[0]
    label_map = {0: 'Negative', 1: 'Offensive', 2: 'Neutral'}
    return {
        'original_text': text,
        'translated_text': translated_text,
        'predicted_label': label_map.get(prediction, f'Class_{prediction}'),
        'probabilities': {label_map.get(i, f'Class_{i}'): prob for i, prob in enumerate(probabilities)}
    }
sample_texts = [
    "This movie is absolutely fire! 🔥",
    "no cap this slaps fr fr 💯",
    "This is so cringe and basic 🙄",
    "bestie this tea is sus lowkey 👀",
    "I'm dead 😂, this party is on fleek!"
]
print("\n" + "="*50)
print("DEMO: Predicting sentiment for Gen-Z texts")
print("="*50)
for text in sample_texts:
    result = predict_sentiment(text)
    print(f"Predicted: {result['predicted_label']}")
    print(f"Probabilities: {result['probabilities']}")
```
